```python
# Import our dependencies
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter

from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python
columns = ['Country', 'Region', 'Happiness Rank', 'Happiness Score'
            'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 
                'Trust (Government Corruption)', 'Generosity' ]


target = ["Happiness Score"]
```

```python
# Load the dataset.
file_path = "2015.csv"
happy_df = pd.read_csv(file_path, index_col=0)
happy_df.head(10)
```

```python
happy_df = happy_df.drop(columns=["Standard Error"], axis=1)
happy_df.head(10)
```

```python
happy_df = happy_df.drop(columns=["Dystopia Residual"], axis=1)
happy_df.head(10)
```

```python
happy_df["Region"].replace({"Western Europe": "10", "North America": "6", "Australia and New Zealand": "1", "Latin America and Caribbean": "4", "Eastern Asia": "3", "Middle East and Northern Africa": "5", "Central and Eastern Europe": "2", "Southeastern Asia": "7", "Southern Asia": "8", "Sub-Saharan Africa": "9" }, inplace=True)
happy_df.head(10)
```

```python
happy_df
```

```python
happy_df.loc[happy_df['Happiness Score'] < 5, 'Happiness Score'] = 0
happy_df.loc[happy_df['Happiness Score'] >= 5, 'Happiness Score'] = 1
happy_df
```

```python
# Create our features
X = happy_df.drop('Happiness Score', axis=1)
X.columns

y = happy_df['Happiness Score']
```

```python
happy_df.columns
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(Counter(y_train))
print(Counter(y_test))
```

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier

brfc = BalancedRandomForestClassifier(n_estimators=100, random_state=1)

brfc.fit(X_train, y_train)
```
